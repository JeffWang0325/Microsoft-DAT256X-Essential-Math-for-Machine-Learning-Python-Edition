```python
# %matplotlib inline
import numpy as np

import matplotlib.pyplot as plt

# 目標函數:y=x^2
def func(x): return np.square(x)

# 目標函數一階導數:dy/dx=2*x
def dfunc(x): return 2 * x

def GD(x_start, df, epochs:int, lr):    
    """ 梯度下降法(GD, Gradient Descent)給定起始點與目標函數的一階導函數，求在epochs次反覆運算中x的更新值
        :param x_start: x的起始點    
        :param df: 目標函數的一階導函數    
        :param epochs: 反覆運算週期    
        :param lr: 學習率(Learning Rate) (i.e. 斜率)  
        :return: x在每次反覆運算後的位置（包括起始點），長度為epochs+1    
    """    
    xs = np.zeros(epochs+1)
    print(xs)
    x = x_start    
    xs[0] = x    
    for i in range(epochs):         
        dx = df(x)        
        # v表示x要改變的幅度        
        v = - dx * lr        
        x += v        
        xs[i+1] = x    
    return xs

# Main
# 超參數 (Hyperparameters): 執行訓練前，可以調整之參數
# 起始權重
x_start = 5 # -5, 5
# 執行週期數
epochs = 20 # 15 20
# 學習率   
lr = 0.3 # 0.01, 0.1 0.3, 0.9, 1.5
# 梯度下降法 
# *** Function 可以直接當參數傳遞 ***
# x = GD(x_start, dfunc, epochs, lr=lr)
x = GD(x_start, dfunc, epochs, lr)
print (x)
# 輸出：[-5.     -2.     -0.8    -0.32   -0.128  -0.0512] for lr = 0.3

color = 'r'
plt.figure(figsize = (10, 8)) 
from numpy import arange
# arange?
t = arange(-6.0, 6.0, 0.01)
# t = arange(-6.0, 7.0)
# t = range(-6, 7)
print(type(t))
print(t)
plt.plot(t, func(t), c='b')
plt.plot(x, func(x), c=color, label='lr={}'.format(lr))    
plt.scatter(x, func(x), c=color, s = 50)    
plt.legend(loc = 'upper left', prop={'size': 20})

plt.xticks(range(-6, 7, 1))
plt.axhline()
plt.axvline()
plt.grid()
plt.show()
```
